```python
import pandas as pd

pre_data = pd.read_csv(r"C:\Users\chand\Downloads\preprocessd_data.csv")
pre_data



from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB,GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,accuracy_score,roc_curve,auc,ConfusionMatrixDisplay
```

```python
X = pre_data.drop("churn",axis=1)
y = pre_data["churn"]

from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=1234)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
GNB = GaussianNB()
GNB.fit(X_train,y_train)
```

```python
KNN = KNeighborsClassifier()
KNN.fit(X_train,y_train)
```

```python
LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
RFC = RandomForestClassifier()
RFC.fit(X_train,y_train)
```

```python
y_pred_dt = dt.predict(X_test)
y_pred_GNB = GNB.predict(X_test)
y_pred_KNN = KNN.predict(X_test)
y_pred_LR = LR.predict(X_test)
y_pred_RFC = RFC.predict(X_test)
```

```python
confusion_matrix_dt = confusion_matrix(y_test,y_pred_dt)
precision_score_dt = precision_score(y_test,y_pred_dt)
recall_score_dt = recall_score(y_test,y_pred_dt)
f1_score_dt= f1_score(y_test,y_pred_dt)
accuracy_score_dt = accuracy_score(y_test,y_pred_dt)


confusion_matrix_GNB = confusion_matrix(y_test,y_pred_GNB)
precision_score_GNB = precision_score(y_test,y_pred_GNB)
recall_score_GNB = recall_score(y_test,y_pred_GNB)
f1_score_GNB= f1_score(y_test,y_pred_GNB)
accuracy_score_GNB = accuracy_score(y_test,y_pred_GNB)


confusion_matrix_KNN = confusion_matrix(y_test,y_pred_KNN)
precision_score_KNN = precision_score(y_test,y_pred_KNN)
recall_score_KNN = recall_score(y_test,y_pred_KNN)
f1_score_KNN= f1_score(y_test,y_pred_KNN)
accuracy_score_KNN = accuracy_score(y_test,y_pred_KNN)


confusion_matrix_LR = confusion_matrix(y_test,y_pred_LR)
precision_score_LR = precision_score(y_test,y_pred_LR)
recall_score_LR = recall_score(y_test,y_pred_LR)
f1_score_LR= f1_score(y_test,y_pred_LR)
accuracy_score_LR = accuracy_score(y_test,y_pred_LR)


confusion_matrix_RFC = confusion_matrix(y_test,y_pred_RFC)
precision_score_RFC = precision_score(y_test,y_pred_RFC)
recall_score_RFC = recall_score(y_test,y_pred_RFC)
f1_score_RFC= f1_score(y_test,y_pred_RFC)
accuracy_score_RFC = accuracy_score(y_test,y_pred_RFC)
```

```python
data = {
    "Decision Tree": [precision_score_dt, recall_score_dt, f1_score_dt, accuracy_score_dt],
    "Naive Bayes": [precision_score_GNB, recall_score_GNB, f1_score_GNB, accuracy_score_GNB],
    "KNN": [precision_score_KNN, recall_score_KNN, f1_score_KNN, accuracy_score_KNN],
    "Logistic Regression": [precision_score_LR, recall_score_LR, f1_score_LR, accuracy_score_LR],
    "Random Forest": [precision_score_RFC, recall_score_RFC, f1_score_RFC, accuracy_score_RFC]
}

metrics_df = pd.DataFrame(data, index=["Confusion Matrix", "Precision", "Recall", "F1 Score", "Accuracy"])

metrics_df
```
