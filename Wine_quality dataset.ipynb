```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_wine
wine = load_wine(as_frame=True)
wine_df=wine["frame"]
wine_df
```

## Divide into X and y

```python
X= wine_df.drop("target", axis=1)
y = wine_df["target"]
```

```python
X
```

```python
y
```

### Vif and Train test split

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = []
for i in range(len(X.columns)):
    vif.append(variance_inflation_factor(X,i))

vif_data = pd.DataFrame(zip(X.columns,vif),columns = ["Features","VIF"])
vif_data
```

```python
con = vif_data["VIF"] < 100
selected_features = vif_data[con]["Features"].to_list()
selected_features
wine_df[selected_features]
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
scaled = sc.fit_transform(wine_df[selected_features])
scaled
```

```python
scaled_data = pd.DataFrame(scaled,columns=selected_features)
wine_df_new = scaled_data
wine_df_new["target"]=wine_df["target"]
wine_df_new
```

```python
X = wine_df_new.drop("target",axis=1)
y = wine_df_new["target"]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

###  ML model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

###  Predictions

```python
y_pred = model.predict(X_test)
y_pred
```

### Evaluation

```python
test_df = X_test.copy()
test_df["Error"] = y_test - y_pred
test_df["Error_Squared"] = test_df["Error"] ** 2
test_df
```

### Save the File
