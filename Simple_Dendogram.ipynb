```python
import numpy as np
import pandas as pd
d = []
val=[[2,5],[3,6],[7,10],[12,10],[6,3],[10,20]]
for i in range(len(val)):
    for j in range(len(val[1:])):
        dis = np.sqrt((val[j][0] - val[i][0])**2 + (val[j][1] - val[i][1])**2)
        d.append([val[i],val[j],dis])
        print(f"The distance between {val[i]} and {val[j]} is {dis}")


df = pd.DataFrame(d,columns = ["A","B","Distance"])
```

# Silhoutte Score

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

mall_df = pd.read_csv(r"C:\Users\chand\Downloads\Mall_Customers.csv")
mall_df
```

```python
x = mall_df[["Annual Income (k$)","Spending Score (1-100)"]].values
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
kmeans = KMeans(n_clusters=5,random_state=1234)
```

```python
kmeans.fit(x)
```

```python
kmeans.labels_
```

```python
silhouette_score(x,kmeans.labels_)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
ss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=(i+1))
    kmeans.fit(x)
    ss.append(silhouette_score(x,kmeans.labels_))
   
```

```python
k_values = range(2,11)
k_values[ss.index(max(ss))]
```

```python
max(ss)
```

```python
k_values = range(2,11)
optimal_k = k_values[np.argmax(ss)]
optimal_silhouette_score = max(ss)

print(optimal_k)
print(optimal_silhouette_score)
```

# Heirarchical Cluster

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

mall_df = pd.read_csv(r"C:\Users\chand\Downloads\Mall_Customers.csv")
mall_df
```

```python
import scipy.cluster.hierarchy as sch
```

```python
X = mall_df.iloc[:,[3,4]].values
```

```python
plt.figure(figsize=(40,30))
dendogram = sch.dendrogram(sch.linkage(X,method="single"))
plt.title("Dendrogram")
plt.xlabel("Clusters")
plt.ylabel("Distance")
plt.show()
```

```python
plt.figure(figsize=(40,30))
dendogram = sch.dendrogram(sch.linkage(X,method="complete"))
plt.title("Dendrogram")
plt.xlabel("Clusters")
plt.ylabel("Distance")
plt.show()
```

```python
plt.figure(figsize=(40,30))
dendogram = sch.dendrogram(sch.linkage(X,method="ward"))
plt.title("Dendrogram")
plt.xlabel("Clusters")
plt.ylabel("Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings("ignore")
hc = AgglomerativeClustering(n_clusters=5,
                            metric="euclidean",
                            linkage="ward")
y_hc = hc.fit_predict(X)
y_hc
```
